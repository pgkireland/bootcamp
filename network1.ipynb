```python
import pandas as pd
import numpy as np

# Create sample network traffic data
data = {
    'Source IP': [f'192.168.1.{i}' for i in range(1, 11)],
    'Destination IP': [f'10.0.0.{i}' for i in range(1, 11)],
    'Protocol': np.random.choice(['TCP', 'UDP', 'ICMP'], size=10),
    'Packets': np.random.randint(50, 500, size=10),
    'Bytes': np.random.randint(1000, 5000, size=10)
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
display(df)
```

```python
# Quick plot: packets per protocol
df.groupby('Protocol')['Packets'].sum().plot(kind='bar', title='Total Packets per Protocol')
```
